```python
import numpy as np
from skopt import gp_minimize
import pandas as pd
import matplotlib.pyplot as plt
import random
import time
```

```python
# Specifying the variables' bound
xmin = -100
xmax = 100
ymin = -100
ymax = 100
amin = -100
amax = 100
bmin = -100
bmax = 100
```

In this package, the parametes are determined by the indices of a single variable. For example, x[0]=x, x[1]=y, x[2]=a, x[3]=b

```python
# specifying the function to be optimized
def func(x):
 
    #2 parameters functions
    trig2 = (np.sin(x[0] * x[1]) - np.cos(5 * x[0]) + np.tan(x[0] + x[1])) 
    power2 = x[0] ** 3 - x[1] ** 2 + x[0] ** 2 - x[1] + 4 
    linear2 = x[0] - x[1]
    #4 parameters functions
    trig4 = np.sin(x[0]*x[1]*x[2]*x[3]) - np.cos(5*x[0]) + np.tan(x[0]/x[3] + x[1]*x[2])
    power4 = (x[2]*x[0])**3 - (x[1]/x[3])**2 + x[0]**2 - (x[3]*x[1]) + 4
    linear4 = (x[2]*x[0]) - (x[1]/x[3])
    return -power4 #---> adding negative sign to the function since this package minimizes the input function 
```

### Data Generator 

Code runs 100 trials to obtain the actual max value of the objective function. This will be used in the next code to calculate the accuracy which is the (calculated max value / actual max value). **This code only needs to be executed if the objective function is changed.**

```python
max_ = []
for i in range(10):
    # random1 = abs(int(np.round(np.random.randint(1, 10) + 1)))
    # random2 = abs(int(np.round(np.random.randint(1, 10) + 1)))
    res = gp_minimize(func,                  # the function to minimize
                  [(-100, 100),(-100, 100),(-100, 100),(-100, 100)],      # the bounds on each dimension of x
                verbose=3,
                  acq_func="LCB",      # the acquisition function
                  n_calls=100,         # the number of evaluations of func
                      acq_optimizer="sampling",
                n_initial_points=50,
                  n_random_starts=50,  # the number of random initialization points
                  noise=0.1**2,      # the noise level (optional)
                  random_state=1234)
    best_x = res.x
    target = res.fun
    # print(res)
    max_.append(target)
    # print(best_x)
max_value = np.max(max_)
```

This code will generate the calculated max value for an optimization of 10 trials. It will divide that by the actual max value to find the accuracy. The mean_time is the average time taken to perform ONE trial. **Run this code everytime any setting of the optimizer is changed** 

```python
Type_of_optimizer = 'Bayes_s'
Type_of_function = 'power'
n_of_params = 4
n_calls=50
itr_time = []
n_of_trials_vector = []
z = []
for i in range(50):
    start = time.time()
    random1 = abs(int(np.round(np.random.randint(13, 25) + 1)))
    random2 = abs(int(np.round(np.random.randint(13, 25) + 1)))
    res = gp_minimize(func,                  # the function to minimize
                  [(-100, 100),(-100, 100),(-100, 100),(-100, 100)],      # the bounds on each dimension of x
                verbose=3,
                  acq_func="LCB",      # the acquisition function
                  n_calls=random1,         # the number of evaluations of f
                      acq_optimizer="sampling",
                n_initial_points=random1,
                  n_random_starts=random1,  # the number of random initialization points
                  noise=0.1**2)       # the noise level (optional)
#                   random_state=1234)
    n_of_trials = random2
    print("num_trials :",n_of_trials)
    best_x = res.x
    target = res.fun
    print(res)
    z.append((func(best_x) / max_value))
    end = time.time()
    timetaken = (end - start)/ n_calls
    itr_time.append(timetaken)
    n_of_trials_vector.append(n_of_trials)
    accuracy = z
    mean_time = itr_time
```

#### Building a dataframe and export it to a csv file

```python
n_of_trials_vector = np.asarray(n_of_trials_vector).reshape(-1,1)
accuracy = np.asarray(accuracy).reshape(-1,1)
mean_time = np.asarray(mean_time).reshape(-1,1)
type_of_opt = np.asarray([Type_of_optimizer]*accuracy.shape[0]).reshape(-1,1)
num_of_params = np.asarray([n_of_params]*accuracy.shape[0]).reshape(-1,1)
type_of_func = np.asarray([Type_of_function]*accuracy.shape[0]).reshape(-1,1)
data1 = np.hstack((n_of_trials_vector, num_of_params, type_of_func,accuracy, mean_time, type_of_opt))
df_skopt_power4_1= pd.DataFrame(data1, columns=['number of trials','number of parameters','type of function','accuracy [calc. max/ actual max]', 'time per trial [s]', 'type_of_opt'])
#print(df_skopt_trig2p)
df_skopt_power4_1.to_csv('skopt_power4_1.csv')
```
