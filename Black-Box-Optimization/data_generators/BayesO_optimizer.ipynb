```python
import numpy as np
from bayeso import bo
import pandas as pd
```

```python
from bayeso.wrappers import wrappers_bo
from bayeso.utils import utils_plotting
import matplotlib.pyplot as plt
from bayeso.utils import utils_bo
import time
```

```python
#Specifying the variables' bound
xmin = -100
xmax = 100
ymin = -100
ymax = 100
# amin = -100
# amax = 100
# bmin = -100
# bmax = 100
```

```python
#specifying the function to be optimized
def func(x):
    
    trig = (np.sin(x[0] * x[1]) - np.cos(5 * x[0]) + np.tan(x[0] + x[1]))
    power = x[0] ** 3 - x[1] ** 2 + x[0] ** 2 - x[1] + 4
    linear = x[0] - x[1]
#     noise = 0 * (np.random.rand(1) - 0.5)
#     fun = np.sin(x[0]*x[1]*x[2]*x[3]) - np.cos(5*x[0]) + np.tan(x[0]/x[3] + x[1]*x[2])
#     power = (x[2]*x[0])**3 - (x[1]/x[3])**2 + x[0]**2 - (x[3]*x[1]) + 4
#     linear = (x[2]*x[0]) - (x[1]/x[3])
#     return -(fun+noise)
    return -linear # adding negative sign to the function since this package minimizes the input function
```

```python
bounds = np.array([[-100, 100],[-100,100]]) #2p
#bounds = np.array([[-100, 100],[-100,100],[-100,100],[-100,100]]) #4p
```

### Data Generator 

Code runs 100 trials to obtain the actual max value of the objective function. This will be used in the next code to calculate the accuracy which is the (calculated max value / actual max value). **This code only needs to be executed if the objective function is changed.**

```python
num_bo1 = 10 # no. of trials
study = bo.BO(bounds, str_optimizer_method_bo= 'L-BFGS-B', debug=False)
max_ = []
for ind_bo in range(0, num_bo1):
    print('BO Round', ind_bo + 1)
    X_final, Y_final, time_final, _, _ = wrappers_bo.run_single_round(
        study,func, num_init=10,num_iter=100,num_samples_ao=100)
    x_best, target = utils_bo.get_best_acquisition_by_history(X_final, Y_final)
    max_value = max_.append(func(x_best))
max_value = np.max(max_)
```

This code will generate the calculated max value for an optimization of 10 trials. It will divide that by the actual max value to find the accuracy. The mean_time is the average time taken to perform ONE trial. **Run this code everytime any setting of the optimizer is changed** 

```python
Type_of_optimizer = 'CMA-ES'
#Type_of_function = 'Trig'
Type_of_function = 'Power'
n_of_params = 2
num_bo = 50
```

```python
model_bo = bo.BO(bounds, debug=False,str_optimizer_method_bo='CMA-ES') # types: 'DIRECT' or 'L-BFGS-B' or 'CMA-ES'
list_X = []
list_Y = []
itr_time = []
n_of_trials_vector = []
z = []
```

```python
for ind_bo in range(0, num_bo):
    start = time.time()
    print('BO Round', ind_bo + 1)
    random1 = abs(int(np.round(np.random.randint(1, 10) + 1)))
    random2 = abs(int(np.round(np.random.randint(1, 10) + 1)))
    X_final, Y_final, time_final, _, _ = wrappers_bo.run_single_round(
        model_bo, func, num_init=random1, num_iter=random2,
        str_initial_method_bo='uniform', str_sampling_method_ao='uniform', num_samples_ao=100)
    n_of_trials = random1 + random2
    bx_best, y_best = utils_bo.get_best_acquisition_by_history(X_final, Y_final)
    list_X.append(bx_best[0])
    list_Y.append(bx_best[1])
    z.append((func(bx_best) / max_value))
    end = time.time()
    timetaken = (end - start) / n_of_trials
    itr_time.append(timetaken)
    n_of_trials_vector.append(n_of_trials)
    accuracy = z
    mean_time = itr_time
```

#### Building a dataframe and export it to a csv file

```python
n_of_trials_vector = np.asarray(n_of_trials_vector).reshape(-1,1)
accuracy = np.asarray(accuracy).reshape(-1,1)
mean_time = np.asarray(mean_time).reshape(-1,1)
type_of_opt = np.asarray([Type_of_optimizer]*accuracy.shape[0]).reshape(-1,1)
num_of_params = np.asarray([n_of_params]*accuracy.shape[0]).reshape(-1,1)
type_of_func = np.asarray([Type_of_function]*accuracy.shape[0]).reshape(-1,1)
data1 = np.hstack((n_of_trials_vector, num_of_params, type_of_func,accuracy, mean_time, type_of_opt))
df_CmaEs_power = pd.DataFrame(data1, columns=['number of trials','number of parameters','type of function','accuracy [calc. max/ actual max]', 'time per trial [s]', 'type_of_opt'])
df_CmaEs.to_csv('BayesO_power_Data_CMA-ES_2P.csv')
```
